```python
import GPUtil
import time
import threading

# Function to monitor GPU usage and log to a file
def monitor_gpu():
    with open("gpu_logs.txt", "w") as log_file:
        while True:
            gpus = GPUtil.getGPUs()
            for gpu in gpus:
                log_entry = f"GPU ID: {gpu.id}, Memory Usage: {gpu.memoryUsed} MB / {gpu.memoryTotal} MB\n"
                log_file.write(log_entry)
                print(log_entry.strip())  # Print to notebook screen for real-time feedback
            log_file.flush()  # Ensure logs are written to the file immediately
            time.sleep(60)

# Start the GPU monitoring in a separate thread
gpu_thread = threading.Thread(target=monitor_gpu)
gpu_thread.daemon = True
gpu_thread.start()

# You can now run other code cells without blocking
print("GPU monitoring is running in the background. Logs are being saved to 'gpu_logs.txt'.")
```
