## Breadth first Search

Breadth-first search (BFS) is an algorithm for traversing or searching tree or graph data 
structures. It starts at the tree root and explores all of the neighbor nodes at the present depth 
prior to moving on to the nodes at the next depth level. It uses a queue (First In First Out) instead 
of a stack and it checks whether a vertex has been discovered before enqueueing the vertex.

### Implementation:

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)

def bfs(G,start,search):
    visited = set()
    queue = Queue()
    queue.enqueue(start)

    while not queue.is_empty():
        
#         print("Starting")
        node = queue.dequeue()
        if node not in visited:
            print("Visited Nodes:",visited)
            if node == search:
                return search
            visited.add(node)
            neighbors = G[node]
            for neighbor in neighbors:
                if neighbor not in visited:
                    queue.enqueue(neighbor)
    return None
#     print("Ending")

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("Search Found: ",bfs(graph, 'A','D'))
```
