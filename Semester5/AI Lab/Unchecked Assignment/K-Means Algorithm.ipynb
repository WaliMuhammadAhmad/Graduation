# K-Means Algorithm

K-means clustering is one of the simplest and popular unsupervised machine learning algorithms. The 
objective of K-means is simple: group similar data points together and discover underlying patterns. To 
achieve this objective, K-means looks for a fixed number (k) of clusters in a dataset. A cluster refers to a 
collection of data points aggregated together because of certain similarities.

## k-Means Clustering code from Scratch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Dataset

```python
dataset = pd.read_csv('iris.csv')
column_names = dataset.columns.tolist()
print("Loaded Data:\n", dataset)
dataset.info()
```

```python
def kmeans(X, y, k, max_iterations=100):
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    
    for _ in range(max_iterations):
        distance = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distance, axis=1)
        
        for i in range(k):
            centroids[i] = np.mean(X[labels == i], axis=0)
            
    return labels, centroids
```

```python
np.random.seed(42)
data = np.concatenate([np.random.normal(loc=5, scale=1, size=(50, 2)),
                       np.random.normal(loc=5, scale=1, size=(50, 2))])

k = 2
labels, centroids = kmeans(data, None, k)
```

### Plot the data points and centroids

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], c='blue', label='Centroids')
plt.title('k-Means Clustering')
plt.xlabel('Features 1')
plt.ylabel('Features 2')
plt.legend()
plt.show()
```

### Predicting and Ploting 

```python
new_data = np.array([[5.0, 5.0]])
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', label='Centroids')
plt.scatter(new_data[0, 0], new_data[0, 1], c='red', marker='x', label='New Data (Predicted)')
plt.title('k-Means Clustering')
plt.xlabel('Features 1')
plt.ylabel('Features 2')
plt.legend()
plt.show()
```
